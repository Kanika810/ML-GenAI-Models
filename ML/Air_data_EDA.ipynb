```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r"air.csv")
```

```python
pollutants = ["PM2.5","PM10","NO2","SO2","CO","O3","NH3"]

for col in pollutants:
    if col in df.columns:
        df[col].fillna(df[col].median())
```

```python
df["AQI"].fillna(df["AQI"].median(), inplace=True)
df["AQI_Bucket"].fillna(df["AQI_Bucket"].mode()[0], inplace=True)
df.head()
```

```python
df = df.sort_values("Date")
df.fillna(method="ffill", inplace=True)
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["AQI"], bins=50, kde=True)
plt.title("Distribution of AQI")
plt.xlabel("AQI")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="AQI_Bucket", data=df)
plt.xticks(rotation=45)
plt.title("AQI Category Distribution")
plt.show()
```

```python
city_aqi = df.groupby("City")["AQI"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
city_aqi.plot(kind="bar")
plt.title("Top 10 Most Polluted Cities")
plt.ylabel("Average AQI")
plt.show()
```

```python
plt.figure(figsize=(12,8))
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Between Pollutants")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="PM2.5", y="AQI", data=df)
plt.title("PM2.5 vs AQI")
plt.show()
```

```python
pollutants = ["PM2.5","PM10","NO2","SO2","CO","O3"]
for col in pollutants:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
plt.figure(figsize=(12,6))
df.groupby("Date")["AQI"].mean().plot()
plt.title("AQI Trend Over Time")
plt.ylabel("AQI")
plt.show()
```
