```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier,export_text,plot_tree
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
data=load_iris()
```

```python
X=data.data
y=data.target
y
```

```python
target_names=data.target_names
class_names=data.target_names
feature_names=data.feature_names
print(f"features:{feature_names}")
print(f"Target_names:{target_names}\n")
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
id3_tree=DecisionTreeClassifier(criterion="entropy", random_state=42)
id3_tree.fit(X_train,y_train)
```

```python
y_pred=id3_tree.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
print(f"Accuracy:{accuracy*100:.2f}%")
```

```python
plt.figure(figsize=(8,9))
plot_tree(id3_tree,feature_names=feature_names,class_names=class_names,filled=True)
plt.title("ID3")
plt.show()
```
