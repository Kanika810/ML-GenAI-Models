```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
data=load_iris()
X=data.data
y=data.target
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
base_classifier=DecisionTreeClassifier()
```

```python
bagging_classifier=BaggingClassifier(base_classifier,n_estimators=10,random_state=42)
bagging_classifier.fit(X_train,y_train)
```

```python
y_pred=bagging_classifier.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
print(f"Accuracy:{accuracy*100:.2f}%")
```

```python
plt.scatter(y_test, y_pred)
plt.title("Actual vs Predicted (Bagging Classifier)")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.show()
```

```python
print("""
This graph compares actual values with predicted values.
- Each point represents a prediction made by the model.
- Points close to the diagonal indicate correct predictions.
- Since most points align well, the model shows very high accuracy.
- The bagging technique improves performance by combining multiple decision trees.
""")
```
