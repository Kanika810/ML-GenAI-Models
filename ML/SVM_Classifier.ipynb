```python
from sklearn.model_selection import cross_val_score,KFold
from sklearn.svm import SVC
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
X,y=iris.data,iris.target
```

```python
svm_classifier=SVC(kernel='linear')
```

```python
num_folds=7
kf=KFold(n_splits=num_folds,shuffle=True,random_state=42)
```

```python
cross_val_results=cross_val_score(svm_classifier,X,y,cv=kf)
cross_val_results
```

```python
print("cross-validation results(Accuracy):")
for i,result in enumerate(cross_val_results,1):
    print(f"Fold(i):{result*100:.2f}%")

print(f" Mean Accuracy:{cross_val_results.mean()*100:.2f}%")
```

```python
from sklearn.model_selection import cross_validate
scoring = ['accuracy', 'precision_macro', 'recall_macro', 'f1_macro']
results = cross_validate(svm_classifier, X, y, cv=5, scoring=scoring)

print("Accuracy:", results['test_accuracy'].mean())
print("Precision:", results['test_precision_macro'].mean())
print("Recall:", results['test_recall_macro'].mean())
print("F1 Score:", results['test_f1_macro'].mean())
```

```python
import matplotlib.pyplot as plt
plt.boxplot(cross_val_results)
plt.title("Accuracy Distribution")
plt.show()
```

```python
print("""
This boxplot shows the accuracy distribution of the SVM model across different cross-validation folds.
- Most accuracy values are close to 1.0, indicating strong model performance.
- The small spread shows the model is stable and consistent.
- One slightly lower value (~0.90) appears as an outlier, meaning the model performed slightly worse in one fold.
- Overall, the SVM model generalizes well with high and reliable accuracy.
""")
```

```python
plt.plot(cross_val_results, marker='o')
plt.title("Cross Validation Accuracy per Fold")
plt.xlabel("Fold")
plt.ylabel("Accuracy")
plt.show()
```

```python
print("""
This line plot shows the accuracy of the SVM model across each cross-validation fold.
- Most folds achieve 100% accuracy, indicating strong performance.
- There is a sharp drop in one fold (~0.90), showing slight variation.
- The model quickly recovers back to high accuracy in the next fold.
- Overall, the model is highly accurate with minor fluctuations across folds.
""")
```
