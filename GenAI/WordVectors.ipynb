```python
!pip install gensim
import gensim.downloader as api
from numpy.linalg import norm

# Load ONE model only
word_vectors = api.load("fasttext-wiki-news-subwords-300")

def explore_word_relationships(word1, word2, word3):
    try:
        # Normalize vectors
        vec1 = word_vectors[word1] / norm(word_vectors[word1])
        vec2 = word_vectors[word2] / norm(word_vectors[word2])
        vec3 = word_vectors[word3] / norm(word_vectors[word3])

        # Vector arithmetic
        result_vector = vec1 - vec2 + vec3

        # (Optional) Normalize result
        result_vector = result_vector / norm(result_vector)

        # Find similar words
        similar_words = word_vectors.similar_by_vector(result_vector, topn=5)

        print(f"\nWord Relationship: {word1} - {word2} + {word3}")
        print("Most similar words:")

        for word, similarity in similar_words:
            print(f"{word}: {similarity:.4f}")

    except KeyError as e:
        print(f"Error: {e} not found in the vocabulary.")
```

```python
explore_word_relationships("king", "man", "woman")
```
